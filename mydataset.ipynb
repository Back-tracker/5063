```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
columns = pd.read_csv("Crop_recommendation.csv");
columns.corr()
```

```python
correlation_matrix = columns.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of crop Dataset')
plt.show()
```

```python
columns[['ph','K']].corr(method='kendall')
```

```python
import numpy as np
print(np.corrcoef(columns['ph'],columns['K']))
```

```python
plt.scatter(columns['ph'],columns['K'],color='red',marker='')
plt.xlabel('ph')
plt.ylabel('K')
plt.title('ph vs K')
sns.scatterplot(x='ph',y='K',data=columns)
plt.grid(True)

plt.show()
```
